Fundamentos de Base de datos para Ciencia de Datos.
==> Una base de datos es una colección de información - usualmente estructurada y/o almacenada en una computadora. el principio básico de ellas es el acceso y capacidad de gestión de datos, actualización, modificación y gracias al uso de éstas, se utiliza para generar modelos y fiabilidad de datos y la usamos como herramienta crítica en la Ciencia de Datos.

==> Data Warehose, es un sistema informático diseñado para almacenar y analizar grandes cantidades de datos estructurados o semiestructurados. Sirve como depósito central, accesible a los usuarios empresariales autorizados que confían en el análisis para tomar decisiones mejor informadas. Un almacén de datos es un componente clave de la mayoría de las estrategias de inteligencia empresarial (BI).

```python
##Importamos Pandas, Numpy y generamos nuestro dataset de la DB de github
##con head() vemos un sumary ràpido de lo que contiene

import pandas as pd
import numpy as np

path="https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"
df=pd.read_csv(path)
df.head()
```

```python
##VERIFICAMOS SI TENEMOS VALORES VACÍOS (empty)
df.isnull().any()
```

```python
##VAMOS A DESCARTAR LOS VALORES VACÍOS
df.dropna(inplace=True)

#comprobamos que no tengamos VACÍOS
print(df.isna().values.any())
df.head()
```

```python
#VAMOS A ELIMINAR LOS VALORES VACÍOS
# 1) primos vamos a copiar el DF a uno nuevo= ndf
ndf= df.copy()
ndf
```

```python
ndf.dropna(thresh = 25, inplace = True)  
ndf

#Con el resultado , el DF pasó de 30K a 28,958 resultados
```

```python
#vamos a ver un summary para ver cómo estamos
ndf.describe()
```

```python
#VEAMOS LOS VALORES ÚNICOS DEL DF.
for columna in ndf:
    i = ndf[columna]
    print('Columna', columna,'    mean: ',i.mean())
```
